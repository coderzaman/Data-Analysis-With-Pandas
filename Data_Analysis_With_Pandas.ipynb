# **Introduction to Pandas üêº**

Pandas is a **powerful** and widely used Python library specifically designed for **data manipulation and analysis**. It provides efficient and easy-to-use data structures, making it an essential tool for working with structured data (like tables or spreadsheets).

---
### **Why Use Pandas?**

Real-world data is often messy. It can contain errors, missing values (nulls), irrelevant information, or outliers. Pandas provides tools to efficiently handle these issues. Its main uses include:

1.  **Data Cleaning:**
    * Identifying and handling **missing values** (e.g., removing or filling them).
    * Removing **irrelevant data** or duplicate entries.
    * Correcting data types.
2.  **Data Preprocessing:**
    * **Transforming data** (e.g., creating new features based on existing ones).
    * **Grouping** and aggregating data for summary statistics.
    * Merging and joining different datasets.
3.  **Data Analysis & Exploration:**
    * Calculating **statistical information** (mean, median, counts, etc.).
    * Identifying **patterns, trends, and outliers**.
    * Filtering and selecting specific subsets of data.

---
In short, Pandas is a fundamental tool for anyone working with data in Python, enabling efficient **cleaning, preprocessing, analysis, and manipulation** tasks.

# **Pandas DataFrame and Series**

Pandas is designed to work with structured data. Its two primary data structures are the **DataFrame** (for 2D tabular data) and the **Series** (for 1D data). We often load data from files like CSV, JSON, or Excel spreadsheets.

### **Reading a CSV and Understanding DataFrame**

First, import the Pandas library (conventionally as `pd`):

```python
# 2. Import Pandas
import pandas as pd
```

Now, let's read a CSV file stored in Google Drive into a Pandas **DataFrame**.

```python
# 3. Read the CSV file
file_path = '/content/drive/MyDrive/Dataset/Pandas_Dt/student_data.csv'
df = pd.read_csv(file_path)

# 4. Display the DataFrame
print("DataFrame (df):")
print(df)

# 5. Check the type
print(f"\nType of df: {type(df)}")
```

**Output:**

```
DataFrame (df):
   StudentID         Name  Age      Grade
0        101        Alice   20         A
1        102          Bob   22         B
2        103      Charlie   21         A
3        104        David   23         C

Type of df: <class 'pandas.core.frame.DataFrame'>
```

As the type shows, `df` is a **DataFrame**. This is Pandas' core structure for representing **2D tabular data**, similar to a spreadsheet or a database table, with rows and columns.

-----

### **Understanding Series**

When you select a **single column** from a DataFrame, you get a Pandas **Series**. A Series is essentially a **1D array** with labels (called an index).

```python
# 6. Access a single column ('StudentID')
std_id_series = df['StudentID']

# 7. Display the Series
print("\nSeries (std_id_series):")
print(std_id_series)

# 8. Check the type
print(f"\nType of std_id_series: {type(std_id_series)}")
```

**Output:**

```
Series (std_id_series):
0    101
1    102
2    103
3    104
Name: StudentID, dtype: int64

Type of std_id_series: <class 'pandas.core.series.Series'>
```

So, a DataFrame is like a container holding one or more Series objects as its columns.

# **Working with Different File Types in Pandas** üìÑ

Pandas is excellent for reading data from various file formats. When you read a supported file type, Pandas automatically converts the data into its primary data structure, the **DataFrame**, allowing you to use all of Pandas' powerful functionalities for manipulation and analysis.

-----

### **Common File Types**

Pandas can handle many file types, including:

1.  **CSV (Comma Separated Values):** Plain text files where values are separated by commas. Very common.
2.  **Excel Files (`.xlsx`, `.xls`):** Microsoft Excel spreadsheets.
3.  **JSON (JavaScript Object Notation):** A lightweight data-interchange format often used in web applications and APIs.
4.  **Parquet:** A columnar storage format optimized for performance, especially with large datasets. Often faster than CSV for read/write operations.
5.  **SQL Databases:** Reading data directly from SQL tables.
6.  **HTML:** Reading tables directly from web pages.
7.  **Plain Text Files (`.txt`):** Though often requires specifying delimiters.

While Pandas supports many formats, **CSV, JSON, and Parquet** are particularly common in data science and machine learning workflows.

-----

### **Reading Files**


# **Pandas Basic Functionalities: Inspecting Your Data**

Once you have loaded your data into a Pandas DataFrame, the first step is always to explore and understand its structure. Pandas provides several essential methods and attributes for this "first look" at your data.

```python
import pandas as pd
import numpy as np

# Load the dataset
df = pd.read_csv('path/to/your/student_data.csv')
```

### `df` vs. `print(df)`

When working in an interactive environment like a Jupyter notebook:

  * **`df` (at the end of a cell):** This displays the DataFrame as a rich, formatted HTML table. It's easier to read.
  * **`print(df)`:** This prints the string representation of the DataFrame, which is less formatted.

If the DataFrame is very large, both methods will truncate the output, showing only the first few and last few rows to save space.

### Viewing Rows: `head()`, `tail()`, and `sample()`

  * **`df.head(n)`:** A **method** that returns the **first `n` rows** (samples) of the DataFrame. The default is 5. This is useful for checking if the data loaded correctly and to see the column headers.

    ```python
    # Show the first 10 rows
    df.head(10)
    ```

  * **`df.tail(n)`:** A **method** that returns the **last `n` rows** of the DataFrame. The default is 5. This is good for checking the end of your data.

    ```python
    # Show the last 10 rows
    df.tail(10)
    ```

  * **`df.sample(n)`:** A **method** that returns a **random sample** of `n` rows from your DataFrame. This is very useful for getting a sense of the data's "typical" values, rather than just the top or bottom.

    ```python
    # Show 10 random rows
    ```

### Getting a Summary: `info()` and `describe()`

These two methods are crucial for data analysis.

  * **`df.info()`:** This **method** provides a concise technical summary of the DataFrame. It is essential for checking:

      * **Total number of entries (rows/samples).**
      * **Total number of columns (features).**
      * **Column names** and their **data types** (`dtype`, e.g., `int64`, `float64`, `object`).
      * **Non-Null Count:** This is the most important part. It shows how many non-missing values are in each column. For example, if you have 150 entries and a column shows "145 non-null," you immediately know that the column has **5 null (or missing) values**.
      * **Memory usage.**

    <!-- end list -->

    ```python
    df.info()
    ```

  * **`df.describe()`:** This **method** automatically calculates descriptive statistics for all **numerical columns** (like `int` and `float`)‚Äîcolumns where calculations are possible. It ignores text (`object`) columns. This is your quickest way to understand the distribution of your data.

      * `count`: Number of non-null values.
      * `mean`: The average.
      * `std`: The standard deviation (how spread out the data is).
      * `min`: The minimum value.
      * `25%`: The 1st quartile (percentile).
      * `50%`: The median (2nd quartile).
      * `75%`: The 3rd quartile.
      * `max`: The maximum value.

    <!-- end list -->

    ```python
    df.describe()
    ```

# **Creating Pandas DataFrames from Python Data Structures**



Pandas makes it easy to convert Python's built-in data structures like lists, tuples, and dictionaries directly into a DataFrame.

```python
import pandas as pd
```

### 1\. From a List of Lists

This method is useful for row-oriented data.

```python
my_list = [['Alice', 25], ['Bob', 30], ['Charlie', 28]]

# Default creation: Columns are 0, 1 and Index is 0, 1, 2
df_list1 = pd.DataFrame(my_list)

# You can specify column and index names during creation
df_list2 = pd.DataFrame(my_list,
                        columns=['Name', 'Age'],
                        index=[1, 2, 3])
```

#### Modifying Columns and Index After Creation

You can also assign columns and index *after* the DataFrame is created.

```python
# Create with default columns/index
df_list3 = pd.DataFrame(my_list)

# Assign new column names
df_list3.columns = ['Name', 'Age']

# Modify the index (e.g., start from 1 instead of 0)
df_list3.index = df_list3.index + 1
```

#### Renaming Columns with `.rename()`

To change specific column names, `.rename()` is very useful.

  * **`inplace=True`:** This modifies the original DataFrame (`df_list3`) directly.
    ```python
    df_list3.rename(columns={'Name': 'Full_Name'}, inplace=True)
    ```
  * **Without `inplace`:** This creates a *new copy* of the DataFrame with the change. You must assign it back to a variable.
    ```python
    df_list3 = df_list3.rename(columns={'Full_Name': 'Name'})
    ```

-----

### 2\. From a List of Tuples

This works identically to a list of lists.

```python
my_tuple_list = (('Alice', 25), ('Bob', 30), ('Charlie', 28))

df_tuple = pd.DataFrame(my_tuple_list, columns=['Name', 'Age'])

# You can use += as a shortcut
df_tuple.index += 1
```

-----

### 3\. From a Dictionary (of Lists)

This is a very common, column-oriented method. The **dictionary keys automatically become the column names**.

```python
my_dict = {
    'Name': ['Alice', 'Bob', 'Charlie'],
    'Age': [25, 30, 28]
}

df_dict = pd.DataFrame(my_dict)
df_dict.index += 1
```

**Important:** When using this method, all the value-lists in the dictionary **must be the same length**. The following code will cause a `ValueError`:

```python
# This will RAISE AN ERROR
# 'Name' has 3 items, but 'Age' only has 2
# my_dict_error = {'Name': ['Alice', 'Bob', 'Charlie'], 'Age': [25, 28]}
# pd.DataFrame(my_dict_error)
```

-----

### 4\. From a List of Dictionaries

This is another powerful and flexible row-oriented method, often used with data from JSON or APIs.

```python
my_list_dict = [
    {'Name': 'Alice', 'Age': 25, 'City': 'Dhaka'},
    {'Name': 'Bob', 'Age': 30, 'City': 'Rajshahi'},
    {'Name': 'Charlie', 'Age': 28} # 'City' is missing
]

df_list_dict = pd.DataFrame(my_list_dict)
```

**Key Feature:** Pandas will automatically create columns for *all* unique keys found in the list. If a specific dictionary (row) is missing a key, Pandas will fill that spot with **`NaN`** (Not a Number), indicating missing data. Notice how Charlie's "City" is `NaN`.

# **Accessing Data with `.loc[]` and `.iloc[]` in Pandas**

A core part of Pandas is selecting specific rows and columns. While you can select columns directly, selecting rows requires special indexers: `.loc` and `.iloc`.

```python
# Assume 'df' is loaded
# df = pd.read_csv('path/to/student_data.csv')
```

### 1\. Accessing Columns

This is the most basic way to select data.

```python
# Access a single column (returns a 1D Pandas Series)
student_ids = df["StudentID"]

# Access multiple columns (returns a 2D DataFrame)
student_details = df[["StudentID", "FullName"]]
```

**Note:** Trying to access a row by its position this way will cause a `KeyError` (e.g., `df[0]` will not work as you expect). This is why we use `.loc` and `.iloc`.

-----

### 2\. Accessing by Label: `.loc[]`

`.loc` selects data based on the **index label** and **column name**.

  * It is **inclusive** on both ends when slicing.
  * Syntax: `df.loc[row_labels, column_labels]`

<!-- end list -->

```python
# Fetch a single row (returns a Series)
row_0 = df.loc[0]

# Fetch multiple specific rows (returns a DataFrame)
rows_2_3_4 = df.loc[[2, 3, 4]]

# Fetch a range of rows
# This is INCLUSIVE: it gets rows labeled 3, 4, 5, 6, and 7
rows_3_to_7 = df.loc[3:7]

# Fetch a single column (returns a Series)
python_marks = df.loc[:, 'Python Marks']

# Fetch multiple columns (returns a DataFrame)
marks_df = df.loc[:, ['Python Marks', 'Algorithm Marks']]

# Fetch specific rows AND columns
# Rows 0-3, and only 'Python' and 'Algorithm' columns
subset = df.loc[0:3, ['Python Marks', 'Algorithm Marks']]
```

-----

### 3\. Conditional Filtering (Boolean Indexing)

This is the most common way to filter data based on values.

```python
# Get all rows where 'Algorithm Marks' are 90 or higher
high_scorers = df[df["Algorithm Marks"] >= 90]

# This can be combined with .loc for more complex selections
high_scorers_names = df.loc[df["Algorithm Marks"] >= 90, "FullName"]
```

-----

### 4\. Changing the Index and Using `.iloc[]`

Sometimes, the default 0, 1, 2... index isn't useful. You might want a meaningful column, like "StudentID", to be the index.

```python
# Method 1: Set the index when reading the file
df = pd.read_csv("student_data.csv", index_col="StudentID")

# Method 2: Set the index after loading
df_original = pd.read_csv("student_data.csv")
# Use inplace=True to modify the original df
# df_original.set_index("StudentID", inplace=True)

# Or, create a new DataFrame
new_df = df_original.set_index("StudentID")
```

#### The Problem this Creates for `.loc[]`

After setting the index to "StudentID", your index labels are no longer `0, 1, 2...`. They are the actual student IDs (e.g., `1001, 1002...`).

```python
# This will now likely cause a KeyError!
# .loc is looking for labels '0', '1', '2', '3' in the "StudentID" index.
# If your StudentIDs start from 101, those labels don't exist.
# new_df.loc[0:4]
```

#### The Solution: `.iloc[]` (Integer-Location Indexing)

`.iloc` selects data based on its **integer position** (from 0), regardless of the index labels.

  * It is **exclusive** on the end when slicing (like standard Python lists).
  * Syntax: `df.iloc[row_positions, column_positions]`

<!-- end list -->

```python

# Fetch the first 4 rows AND the columns at position 1 and 2
# Rows 0,1,2,3 (exclusive) and Columns 1,2 (exclusive)
```

# **Dropping Rows/Columns, Assigning Values, and Iteration**

This note covers how to modify a DataFrame by removing data, changing existing data, and iterating over rows.

### 1\. Dropping Rows and Columns with `.drop()`

The `.drop()` method is used to remove rows or columns.

  * **Drop a Single Row:**
    To remove a single row, pass its **index label**.

    ```python
    # Drops the row with index label 1
    df.drop(1, inplace=True)
    ```

  * **Drop Multiple Rows:**
    Pass a list of index labels to drop multiple rows at once.

    ```python
    # Drops the rows with index labels 1, 2, and 3
    df.drop([1, 2, 3], inplace=True)
    ```

  * **Drop a Column:**
    To drop a column, you must specify `axis=1` (or `axis='columns'`) to tell Pandas you are targeting a column label, not a row label.

    ```python
    # axis=1 signifies that we are dropping a column
    df.drop('Instructor', axis=1, inplace=True)
    ```

### 2\. Resetting the Index After Drops

After dropping rows, your index will have gaps (e.g., 0, 2, 3, 5...). To re-order the index sequentially, use `.reset_index()`.

  * The **`drop=True`** parameter is essential. It discards the old, gapped index. If you forget it, the old index will be added as a *new column* to your DataFrame.

<!-- end list -->

```python
# Re-creates a clean index from 0 to n-1
df.reset_index(drop=True, inplace=True)
```

### 3\. Assigning and Modifying Values with `.loc[]`

The recommended way to change values is by using `.loc[]` to select the data you want to change and then assigning a new value.

  * **Set a Single Cell's Value:**
    Provide the `[row_label, column_label]`.

    ```python
    df.loc[3, 'Python Marks'] = 90
    df.loc[3, 'CompletionStatus'] = 'Completed'
    ```

  * **Modify Multiple Values (Vectorization):**
    You can select a "slice" of the DataFrame and apply an operation to all of it at once. This is very fast and efficient.

    ```python
    # Adds 2 to all values in the specified rows and columns
    df.loc[0:3, ['Python Marks', 'Algorithm Marks']] += 2
    ```

### 4\. Iteration (Use Sparingly\!) ‚ö†Ô∏è

While you *can* loop over a DataFrame, **it is strongly discouraged**. It is very slow and inefficient compared to vectorized operations (like the `+= 2` example above).

If you absolutely must loop, you have two main options:

  * **`.iterrows()`:**
    This method loops over each row, returning the index (`i`) and the row's data as a Pandas `Series` (`series`).

    ```python
    for i, series in df.iterrows():
        print(f"Index: {i}")
        print(series['FullName']) # Access data like a Series
    ```

  * **`.itertuples()`:**
    This method is generally **much faster** than `.iterrows()`. It returns each row as a lightweight `namedtuple`, which is more efficient.

    ```python
    # index=False stops the index from being added as the first element of the tuple
    for row_tuple in df.itertuples(index=False):
        print(row_tuple)
        print(row_tuple.FullName) # Access data by name
    ```

# **Sorting a DataFrame with `.sort_values()`**

You can easily reorder your DataFrame based on the values in one or more columns using the `.sort_values()` method. By default, this method creates a **new copy** of the DataFrame; it does not modify the original unless you use `inplace=True`.

```python
# Create a copy to see the sorted results
copy_df = df.sort_values(by='Algorithm Marks')
```

### 1\. Single Column Sorting

  * **Ascending Order (Default):**
    This is the default behavior. It sorts from smallest to largest.

    ```python
    # Sorts by 'Algorithm Marks' from lowest to highest
    copy_df = df.sort_values(by='Algorithm Marks')
    ```

  * **Descending Order:**
    To sort from largest to smallest, set `ascending=False`.

    ```python
    # Sorts by 'Algorithm Marks' from highest to lowest
    copy_df = df.sort_values(by='Algorithm Marks', ascending=False)
    ```

### 2\. Multi-Column Sorting

You can sort by multiple columns by passing a **list of column names**.

  * **Same Order (All Descending):**
    Pandas sorts by the *first column in the list*, then uses the *second column to break ties* (and so on).

    ```python
    # 1. First, sorts by 'Data Structure Marks' (descending)
    # 2. If any 'Data Structure Marks' are the same, it sorts those rows by 'Python Marks' (descending)
    copy_df = df.sort_values(by=['Data Structure Marks', 'Python Marks'], ascending=False)
    ```

  * **Mixed Order (Ascending and Descending):**
    To set different sort orders for each column, pass a **list of booleans** to the `ascending` parameter. The order of the list must match the order of your columns.

      * `0` or `False` = Descending
      * `1` or `True` = Ascending

    <!-- end list -->

    ```python
    # Sorts by 'Data Structure Marks' (ascending=False -> Descending)
    # Then sorts ties by 'Python Marks' (ascending=True -> Ascending)
    copy_df = df.sort_values(by=['Data Structure Marks', 'Python Marks'], ascending=[0, 1])
    ```

# **Filtering Data Based on Conditions (Boolean Indexing)**

One of the most powerful features of Pandas is **Boolean indexing**, which allows you to filter a DataFrame based on one or more conditions. This creates a "mask" of `True`/`False` values and selects only the rows where the condition is `True`.

You can use either the standard `df[]` brackets or `df.loc[]` for this.

### 1\. Single Condition Filtering

You create a condition, and Pandas returns all rows that match.

```python
# Approach 1: Using df.loc[]
# Find students whose CompletionStatus is "Not Started"
not_started_df = df.loc[df['CompletionStatus'] == "Not Started"]

# Approach 2: Using df[] (This is the more common, shorter way)
# Find students whose CompletionStatus is "Completed"
completed_df = df[df['CompletionStatus'] == "Completed"]
```

### 2\. Multiple Condition Filtering

To combine conditions, you must use:

  * `&` for **AND**
  * `|` for **OR**
  * `~` for **NOT**

**CRITICAL:** Each individual condition **must** be wrapped in parentheses `( )`.

```python
# Example of AND (&)
# Find students whose status is "Completed" AND Data Structure Marks >= 85
filtered_df = df[ (df['CompletionStatus'] == "Completed") & (df['Data Structure Marks'] >= 85) ]

# This also works identically with .loc[]
filtered_df_loc = df.loc[ (df['CompletionStatus'] == "Completed") & (df['Data Structure Marks'] >= 85) ]
```
